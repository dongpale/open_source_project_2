```python
import pandas as pd

# 구글 코랩 환경에서 하여 경로 설정을 아래와 같이 설정함.
df = pd.read_csv('/content/sample_data/2019_kbo_for_kaggle_v2.csv')

## 2-1.

for year in range(2015, 2019):
  print("\n{}년 ".format(year))

  ## 연도 추출
  extract_year = df[df['year'] == year]

  for category in ['H', 'avg', 'HR', 'OBP']:
    print("\nTop 10 Players in {}".format(category))

    ## 연도별로 안타, 타율, 홈런, 출루율에 대해 내림차순으로 정렬
    sorted_year = extract_year.sort_values(by=category, ascending=False)
    ## 내림차순으로 정렬된 데이터를 기반으로 안타, 타율, 홈런, 출루율에 대해서 순위 지정
    sorted_year['rank'] = sorted_year[category].rank(ascending=False, method='average')

    top_10_players = sorted_year[sorted_year['rank'] <= 10]
    print(top_10_players['batter_name'])

print('\n')
## 2-2.

print("2018년도 포지션별 승리 기여도 가장 높은 선수들")
## 2018년도 포수 추출
extract_2018_year_catcher = df[(df['year'] == 2018) & (df['cp'] == "포수")]
## 포수 중 승리 기여도가 가장 높은 인덱스 추출
max_catcher= extract_2018_year_catcher['war'].idxmax()
## 포수 중 승리 기여도 가장 높은 행을 result_catcher 변수에 저장 후 선수 이름 출력
result_catcher = extract_2018_year_catcher.loc[max_catcher]
print("포수 : ", result_catcher['batter_name'])


## 2018년도 1루수 추출
extract_2018_year_first = df[(df['year'] == 2018) & (df['cp'] == "1루수")]
## 1루수 중 승리 기여도가 가장 높은 인덱스 추출
max_first= extract_2018_year_first['war'].idxmax()
## 1루수 중 승리 기여도 가장 높은 행을 result_first 변수에 저장 후 선수 이름 출력
result_first = extract_2018_year_first.loc[max_first]
print("1루수 : ", result_first['batter_name'])

## 2018년도 2루수 추출
extract_2018_year_second = df[(df['year'] == 2018) & (df['cp'] == "2루수")]
## 2루수 중 승리 기여도가 가장 높은 인덱스 추출
max_second= extract_2018_year_second['war'].idxmax()
## 2루수 중 승리 기여도 가장 높은 행을 result_second 변수에 저장 후 선수 이름 출력
result_second = extract_2018_year_second.loc[max_second]
print("2루수 : ", result_second['batter_name'])

## 2018년도 3루수 추출
extract_2018_year_third = df[(df['year'] == 2018) & (df['cp'] == "3루수")]
## 3루수 중 승리 기여도가 가장 높은 인덱스 추출
max_third= extract_2018_year_third['war'].idxmax()
## 3루수 중 승리 기여도 가장 높은 행을 result_third 변수에 저장 후 선수 이름 출력
result_third = extract_2018_year_third.loc[max_third]
print("3루수 : ", result_third['batter_name'])

## 2018년도 유격수 추출
extract_2018_year_shortstop = df[(df['year'] == 2018) & (df['cp'] == "유격수")]
## 유격수 중 승리 기여도가 가장 높은 인덱스 추출
max_shortstop= extract_2018_year_shortstop['war'].idxmax()
## 유격수 중 승리 기여도 가장 높은 행을 result_shortstop 변수에 저장 후 선수 이름 출력
result_shortstop = extract_2018_year_shortstop.loc[max_shortstop]
print("유격수 : ", result_shortstop['batter_name'])

## 2018년도 좌익수 추출
extract_2018_year_left = df[(df['year'] == 2018) & (df['cp'] == "좌익수")]
## 좌익수 중 승리 기여도가 가장 높은 인덱스 추출
max_left= extract_2018_year_left['war'].idxmax()
## 좌익수 중 승리 기여도 가장 높은 행을 result_left 변수에 저장 후 선수 이름 출력
result_left = extract_2018_year_left.loc[max_left]
print("좌익수 : ", result_left['batter_name'])

## 2018년도 중견수 추출
extract_2018_year_center = df[(df['year'] == 2018) & (df['cp'] == "중견수")]
## 중견수 중 승리 기여도가 가장 높은 인덱스 추출
max_center= extract_2018_year_center['war'].idxmax()
## 중견수 중 승리 기여도 가장 높은 행을 result_center 변수에 저장 후 선수 이름 출력
result_center = extract_2018_year_center.loc[max_center]
print("중견수 : ", result_center['batter_name'])

## 2018년도 우익수 추출
extract_2018_year_right = df[(df['year'] == 2018) & (df['cp'] == "우익수")]
## 우익수 중 승리 기여도가 가장 높은 인덱스 추출
max_right= extract_2018_year_right['war'].idxmax()
## 우익수 중 승리 기여도 가장 높은 행을 result_right 변수에 저장 후 선수 이름 출력
result_right = extract_2018_year_right.loc[max_right]
print("우익수 : ", result_right['batter_name'])

print('\n')
## 2-3.

## 비교할 득점, 안타, 홈런, 타점, 도루, 승리 기여도, 타율, 출루율, 장타율에 대해서 리스트 변수에 저장
compare_columns = ['R', 'H', 'HR', 'RBI', 'SB', 'war', 'avg', 'OBP', 'SLG']
## 연봉과 compare_columns 안에 있는 것들 간의 상관 관계들을 계산
correlations = df[compare_columns].corrwith(df['salary'])
## 연봉과 상관관계가 가장 높은 열 추출 후 해당 열의 이름을 출력
max_corr_column = correlations.idxmax()
print("The column with the highest correlation with 'salary' is: {}".format(max_corr_column))
```
